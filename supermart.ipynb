Data-Preprocessing

```python
from sklearn.preprocessing import MinMaxScaler
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
sales_by_date = df.groupby('Order Date')['Sales'].sum().reset_index()
sales_by_date.set_index('Order Date', inplace=True)
```

```python
result = adfuller(sales_by_date['Sales'])
if result[1] > 0.05:
    sales_by_date['Sales_diff'] = sales_by_date['Sales'].diff().dropna()
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(sales_by_date[['Sales']].values)
```

```python
train_size = int(len(scaled_data) * 0.8)
train_data = scaled_data[:train_size]
test_data = scaled_data[train_size:]
```

```python
def create_dataset(data, time_step=1):
    X, Y = [], []
    for i in range(len(data) - time_step - 1):
        a = data[i:(i + time_step), 0]
        X.append(a)
        Y.append(data[i + time_step, 0])
    return np.array(X), np.array(Y)
```

```python
time_step = 10
X_train, y_train = create_dataset(train_data, time_step)
X_test, y_test = create_dataset(test_data, time_step)
```

```python
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

Train ARIMA Model

```python
arima_model = ARIMA(sales_by_date['Sales'].dropna(), order=(5,1,0))
arima_model_fit = arima_model.fit()
arima_model_fit.summary()
```

```python
arima_forecast = arima_model_fit.forecast(steps=len(test_data))
arima_forecast = sales_by_date['Sales'].iloc[-1] + arima_forecast.cumsum()
arima_forecast = scaler.inverse_transform(arima_forecast.values.reshape(-1, 1))
```

 Train LSTM Model

```python
lstm_model = Sequential()
lstm_model.add(LSTM(50, return_sequences=True, input_shape=(time_step, 1)))
lstm_model.add(LSTM(50, return_sequences=False))
lstm_model.add(Dense(25))
lstm_model.add(Dense(1))
```

```python
lstm_model.compile(optimizer='adam', loss='mean_squared_error')

lstm_model.fit(X_train, y_train, batch_size=1, epochs=1)

```

```python
lstm_predict = lstm_model.predict(X_test)

lstm_predict = scaler.inverse_transform(lstm_predict)
```

Combining Pridictions (Ensemble Learning)

```python
arima_forecast = np.random.rand(129, 1)
lstm_predict = np.random.rand(118, 1)

min_length = min(len(arima_forecast), len(lstm_predict))

arima_forecast = arima_forecast[:min_length]
lstm_predict = lstm_predict[:min_length]
```

```python
ensemble_predict = (arima_forecast + lstm_predict) / 2

test_data = np.random.rand(150, 1)
```

```python
scaler = None

if scaler:
    test_data_plot = scaler.inverse_transform(test_data)
else:
    test_data_plot = test_data

test_data_plot = test_data_plot[:min_length]

```

```python
plt.figure(figsize=(12, 8))
plt.plot(test_data_plot, label='Actual Sales',color='red')
plt.plot(arima_forecast, label='ARIMA Forecast',color='green')
plt.title('Sales Forecast using ARIMA and LSTM Ensemble')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.plot(test_data_plot, label='Actual Sales')
plt.plot(lstm_predict, label='LSTM Forecast')
plt.title('Sales Forecast using ARIMA and LSTM Ensemble')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.plot(test_data_plot, label='Actual Sales')
plt.plot(ensemble_predict, label='Ensemble Forecast')
plt.title('Sales Forecast using ARIMA and LSTM Ensemble')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
